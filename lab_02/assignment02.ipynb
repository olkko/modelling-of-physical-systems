In the task I repeated plot from lecture 1 slide â„–13. The model is based on one dimensional advection diffusion equation. The function determines the concentration *c(x, t)* at any given position *x* and time *t*.

$c(x, t) = \frac{M}{A} \frac{1}{2\sqrt{\pi D}}\frac{x}{t^\frac{3}{2}}\exp{[-\frac{(ut-x)^2}{4Dt}]}$

```python
import pint
import numpy as np
from matplotlib import pyplot as plt
from open_atmos_jupyter_utils import show_plot
```

```python
si = pint.UnitRegistry()
M = 0.205*si.kg                        # mass
A = 0.4 * (si.m*si.m*si.m)/si.s        # area
D = 0.5 * (si.m*si.m)/si.s             # difussion
u = 1.35 * si.m/si.s                   # Velosity
c = 1*si.kg / (si.m*si.m*si.m)
```

```python
x = np.linspace(30, 100, 100) * si.m
t = np.linspace(0.1, 100, 100) * si.s
```

```python
c = (M / A) * (1.0 / (2.0 * np.sqrt(np.pi * D))) * (x / (np.power(t, 1.5))) * np.exp((-1)*(np.power((u * t - x), 2)) / (4.0 * D * t))
```

```python
si.setup_matplotlib()
def plot(fig):
    fig.gca().plot(t, c, label='Concentration at given position $x$ and time $t$')
    fig.legend()
    fig.gca().grid()
```

```python
fig = plt.figure(figsize=(6,4))
plot(fig)
show_plot(fig=fig)
```
